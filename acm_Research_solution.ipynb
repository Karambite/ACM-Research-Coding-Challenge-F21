```python
#import statements
import numpy as np
import nltk
nltk.download(['vader_lexicon','stopwords','punkt'])
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import re
from textblob import TextBlob
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import string
```

```python
# creating the stemmer and pre modeled sentiment analyzer objects
sid = SentimentIntensityAnalyzer()
ps = PorterStemmer()

# store the list of stop words from nltk
stopwords = nltk.corpus.stopwords.words("english")

# open the text file
data = open("input.txt")

# replace all spaces and line breaks with white spaces
t = data.read().lower().replace('\n', ' ')

# --- Filtering the text by sentence ---

#split the passage into sentences
sentences = nltk.sent_tokenize(t)

for w in range(len(sentences)):
  sentences = [w for w in sentences if w not in stopwords]

# filter and process the text to analyze
for x in range(len(sentences)):
  sentences[x] = re.sub('\[.*?\]',' ',sentences[x])
  sentences[x] = re.sub('[%s]'%re.escape(string.punctuation),' ',sentences[x])
  sentences[x] = re.sub('\w*\d\w*',' ',sentences[x])
  sentences[x] = re.sub('[''"",,,]',' ',sentences[x])

# stem the filtered words into new array
processed_words = []
for y in range(len(sentences)):
    processed_words.append(ps.stem(sentences[y]))

text = ""
text = text.join(processed_words)

#store the sentiment scores of the sentences
sentiment_scores = []
for z in range(len(processed_words)):
    sentiment_scores.append(sid.polarity_scores(processed_words[z])['compound'])
  
#store the subjectivity scores of the sentences
subjectivity_scores = []
for z in range(len(processed_words)):
    subjectivity_scores.append(TextBlob(processed_words[z]).sentiment.subjectivity)

# creating a dataframe to store all the processed sentences and their respective sentiment scores
df = pd.DataFrame({'Sentence' : processed_words, 'Sentiment Score' : sentiment_scores, 'Subjectivity Score' : subjectivity_scores})
left_aligned_df = df.style.set_properties(**{'text-align': 'left'})
display(left_aligned_df)

# Final analysis of the text
sid_value = sid.polarity_scores(str(text))['compound']
subjectivity_value = TextBlob(text).sentiment.subjectivity
sid_result = "Neutral" if sid_value < .1 and sid_value > -.1 else ("Positive" if sid_value > 0 else "Negative")
subjectivity_result = "Subjective" if subjectivity_value > .5 else "Not Subjective"

# display the results
print("The Sentiment score of the text is " + str(sid_value))
print("This means that the text has a " + sid_result)
print("The Subjectivity score of the text is " + str(subjectivity_value))
print("This means that the text is " + subjectivity_result)

```
